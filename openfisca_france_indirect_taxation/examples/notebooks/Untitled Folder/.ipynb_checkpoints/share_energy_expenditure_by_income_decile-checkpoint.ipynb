# Percentage of households' energy expenditures in total income

# Import general modules

```python
from __future__ import division

import pandas
import seaborn
```

# Import modules specific to OpenFisca

```python
from openfisca_france_indirect_taxation.examples.utils_example import graph_builder_line_percent
from openfisca_france_indirect_taxation.surveys import SurveyScenario
from openfisca_france_indirect_taxation.almost_ideal_demand_system.aids_estimation_from_stata import get_elasticities
from openfisca_france_indirect_taxation.examples.calage_bdf_cn_energy import get_inflators_by_year_energy
```

# Import new colors

```python
seaborn.set_palette(seaborn.color_palette("Set2", 12))
%matplotlib inline
```

# Simulate energy's expenditure share by income decile

```python
# Slection des variables utilises pour la simulation

inflators_by_year = get_inflators_by_year_energy(rebuild = False)
year = 2012
data_year = 2011
elasticities = get_elasticities(data_year)
inflation_kwargs = dict(inflator_by_variable = inflators_by_year[year])
del inflation_kwargs['inflator_by_variable']['somme_coicop12']

simulated_variables = ['depenses_energies', 'depenses_energies_logement', 'depenses_essence', 'depenses_diesel', 'poste_coicop_722',
                       'poste_coicop_451', 'poste_coicop_452', 'poste_coicop_453', 'poste_coicop_454',
                       'rev_disp_loyerimput', 'somme_coicop12']

survey_scenario = SurveyScenario.create(
    elasticities = elasticities,
    inflation_kwargs = inflation_kwargs,
    reform_key = 'taxe_carbone',
    year = year,
    data_year = data_year
    )
pivot_table = pandas.DataFrame()
for values in simulated_variables:
    pivot_table = pandas.concat([
        pivot_table,
        survey_scenario.compute_pivot_table(values = [values], columns = ['niveau_vie_decile'])
        ])
df = pivot_table.T

df.rename(columns = {'rev_disp_loyerimput': 'disposable income',
    'somme_coicop12': 'total expenditures'},
    inplace = True)

for element in [u'disposable income', u'total expenditures']:
    df['Energy share in {}'.format(element)] = df['depenses_energies'] / df['{}'.format(element)]
    df['Housing energy share in {}'.format(element)] = df['depenses_energies_logement'] / df['{}'.format(element)]
    df['Transport energy share in {}'.format(element)] = df['poste_coicop_722'] / df['{}'.format(element)]
    df['Diesel share in {}'.format(element)] = df['depenses_diesel'] / df['{}'.format(element)]
    df['Gasoline share in {}'.format(element)] = df['depenses_essence'] / df['{}'.format(element)]
    df['Electricity share in {}'.format(element)] = df['poste_coicop_451'] / df['{}'.format(element)]
    df['Natural gas share in {}'.format(element)] = df['poste_coicop_452'] / df['{}'.format(element)]
    df['Domestic fuel share in {}'.format(element)] = df['poste_coicop_453'] / df['{}'.format(element)]
    df['Solid fuels share in {}'.format(element)] = df['poste_coicop_454'] / df['{}'.format(element)]
    
    # Ralisation de graphiques
    print 'Percentage of energy expenditure in {} :'.format(element)
    graph_builder_line_percent(df[['Energy share in {}'.format(element), 'Housing energy share in {}'.format(element),
                                   'Transport energy share in {}'.format(element)]], 1, 1)
    graph_builder_line_percent(df[['Diesel share in {}'.format(element), 'Gasoline share in {}'.format(element)]], 1, 1)
    graph_builder_line_percent(df[['Electricity share in {}'.format(element), 'Natural gas share in {}'.format(element),
                                  'Domestic fuel share in {}'.format(element), 'Solid fuels share in {}'.format(element)]], 1, 1)
```
